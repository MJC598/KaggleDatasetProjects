```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(1)
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
random_vals_x = np.random.choice(x, size=30)
random_vals_y = np.random.choice(y, size=30)
random_vals_z = 0.12 * random_vals_x + 0.01 * random_vals_y + rng.uniform(low=-0.75, high=2.25, size=30)

x, y = np.meshgrid(x, y)
eq = 0.12 * x + 0.01 * y + 1.09


plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.scatter(random_vals_x, random_vals_y, random_vals_z)
ax.plot_surface(x, y, eq)
plt.show()
```
